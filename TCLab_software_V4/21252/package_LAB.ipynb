```python
import numpy as np
import matplotlib.pyplot as plt
from package_LAB import lead_lag  # Import du package

# Définition des paramètres
K = 1.5   # Gain
T_lead = 2.0  # Constante de temps du numérateur
T_lag = 5.0   # Constante de temps du dénominateur
Ts = 0.1  # Temps d'échantillonnage

# Génération d'une entrée en échelon
t = np.arange(0, 50, Ts)  # Temps
MV = np.ones_like(t)  # Signal d'entrée (échelon)

# Application de la fonction Lead-Lag
PV = lead_lag(MV, K, T_lead, T_lag, Ts)

#  Tracé des courbes
plt.figure(figsize=(10, 5))
plt.plot(t, MV, 'b-', label="Entrée MV (Echelon)")
plt.plot(t, PV, 'r-', label="Sortie PV (Lead-Lag)")
plt.xlabel("Temps [s]")
plt.ylabel("Valeur")
plt.title("Test du Modèle Lead-Lag")
plt.legend()
plt.grid()
plt.show()
```
